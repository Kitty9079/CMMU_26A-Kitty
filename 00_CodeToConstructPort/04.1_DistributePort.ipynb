```python
import os
import pandas as pd
```

```python
# Directory containing the workbooks
workbook_folder = r"C:\Users\Kantinant Sukkert\Desktop\Code\ThematicTesting\Test_MY\input"  # Update with your folder path
output_folder = r"C:\Users\Kantinant Sukkert\Desktop\Code\ThematicTesting\Test_MY\output"  # Update with the path to save output files
```

```python
def Distributed(workbook_folder):
    # Loop over all Excel files in the folder
    for workbook in os.listdir(workbook_folder):
        if workbook.endswith('.xlsx'):
            # Construct the full file path
            file_path = os.path.join(workbook_folder, workbook)
        
            # Load the Excel file
            excel_file = pd.ExcelFile(file_path)
        
            # Process each sheet in the workbook
            for sheet_name in excel_file.sheet_names:
                # Read the data from the sheet
                df = pd.read_excel(file_path, sheet_name=sheet_name)

                # Format the sheet name: convert '12-2009' to '2009-12'
                if '-' in sheet_name:
                    month, year = sheet_name.split('-')
                    formatted_sheet_name = f"{year}-{month}"
                else:
                    formatted_sheet_name = sheet_name  # Keep the sheet name unchanged if no '-'
            
                # Format the sheet name: replace '-' with '_' and prefix with 'PORT_'
                formatted_sheet_name = "PORT_" + formatted_sheet_name.replace('-', '_')
            
                # Construct the output file path
                output_file = os.path.join(output_folder, f"{formatted_sheet_name}.csv")
            
                # Save the DataFrame to a new Excel file
                df.to_csv(output_file, index=False)
            
                print(f"Saved sheet '{sheet_name}' from workbook '{workbook}' to file '{output_file}'")
```

```python
Distributed(workbook_folder)
```
