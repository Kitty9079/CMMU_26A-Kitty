```python
import os
import pandas as pd

# Define the directories
stat_path = r"C:\Users\Kantinant Sukkert\Desktop\BAC Thematic\1_Input\04_Stat_ROE\03_Stock_3YStat_Only"
roi_path = r"C:\Users\Kantinant Sukkert\Desktop\BAC Thematic\1_Input\04_Stat_ROE\02_Stock_ME_ROI_Lag"
output_file = r"C:\Users\Kantinant Sukkert\Desktop\BAC Thematic\2_Output\04_MergedStockFile\AllMeragedStock.csv"

# Function to extract the number before the stock name
def extract_number(filename):
    return int(filename.split('_')[0])

# Get the list of files in the stat directory and sort them by the number before the stock name
sorted_csv_files = sorted(os.listdir(stat_path), key=extract_number)

# List to store each merged DataFrame
merged_list = []

# Loop through each sorted CSV file in the stat directory
for filename in sorted_csv_files:
    if filename.endswith(".csv"):
        # Construct full file paths
        stat_file = os.path.join(stat_path, filename)
        roi_file = os.path.join(roi_path, filename)
        
        # Check if the corresponding roi file exists
        if os.path.exists(roi_file):
            # Read the stat and roi CSV files
            stat_df = pd.read_csv(stat_file)
            roi_df = pd.read_csv(roi_file)
            
            # Merge the data on 'Month_Year'
            merged_df = pd.merge(stat_df, roi_df[['Month_Year', 'pct_ROI', 'ROI_Lag1']], 
                                 on='Month_Year', how='left')
            
            # Append the merged DataFrame to the list
            merged_list.append(merged_df)
            print(f"Merged {filename} successfully.")
        else:
            print(f"ROI file for {filename} not found in {roi_path}")

# Concatenate all merged DataFrames into one
all_merged_df = pd.concat(merged_list, ignore_index=True)

# Save the concatenated DataFrame to a single CSV file
all_merged_df.to_csv(output_file, index=False)
print(f"All merged data saved to {output_file}")
```
