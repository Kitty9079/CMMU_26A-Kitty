```python
import pandas as pd


df_sample = pd.read_csv(r"C:\Users\Kantinant Sukkert\Desktop\Code\ThematicTesting\Test_MY\Testoutputport\PORT_2012_3.csv")

# Calculate the quintile thresholds
quintile_thresholds_sample = df_sample['volatility_stock'].quantile([0.2, 0.4, 0.6, 0.8])

# Assign stocks to quintiles based on volatility in ascending order
df_sample['quintile'] = pd.qcut(df_sample['volatility_stock'], q=5, labels=[1, 2, 3, 4, 5])

# Count the number of stocks per quintile
stock_counts_per_quintile_sample = df_sample['quintile'].value_counts().sort_index()

print("Quintile Thresholds:")
print(quintile_thresholds_sample)

print("\nStock Counts per Quintile:")
print(stock_counts_per_quintile_sample)
```

## Remark

stock_counts_per_quintile_sample = df_sample['quintile'].value_counts().sort_index()

df_sample['quintile'].value_counts(): This counts how many stocks fall into each quintile.
sort_index(): We sort the counts by the quintile label so that they are ordered from 1 to 5.

df_sample['quintile'] = pd.qcut(df_sample['volatility_stock'], q=5, labels=[1, 2, 3, 4, 5])

pd.qcut: This method divides the volatility_stock values into q equal parts (quintiles in this case, so q=5).
labels=[1, 2, 3, 4, 5]: We label these quintiles as 1, 2, 3, 4, and 5, with 1 being the lowest volatility quintile and 5 being the highest volatility quintile.
A new column, quintile, is added to df_sample, indicating the quintile to which each stock belongs

```python
import pandas as pd
import os
import glob

# Define the folder path containing the CSV files
folder_path =  r"C:\Users\Kantinant Sukkert\Desktop\BAC Thematic\02_Output\11_Volatility_Wight_BAV_Summary\BAV_Summary"

# Define the file pattern to search for all CSV files in the folder
file_pattern = os.path.join(folder_path, "*.csv")

# Get all matching CSV files
csv_files = glob.glob(file_pattern)

# Create an empty list to store dataframes
df_list = []

# Loop through each CSV file and append its content to the list
for file in csv_files:
    try:
        df = pd.read_csv(file)
        df_list.append(df)
        print(f"Successfully read: {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Check if any data was read
if df_list:
    # Concatenate all dataframes into one
    combined_df = pd.concat(df_list, ignore_index=True)

    # Save the combined dataframe to a new CSV file
    output_file = os.path.join(folder_path, "combined_summary.csv")
    combined_df.to_csv(output_file, index=False)

    print(f"All files have been combined into {output_file}")
else:
    print("No CSV files found or successfully read in the specified folder.")
```
